```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
mnist = tf.keras.datasets.mnist
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()
```

```python
print("Train:",train_images.shape, "Test:", test_images.shape)
```

```python
plt.imshow(train_images[1000], cmap="grey")
plt.colorbar()
```

```python
train_images = train_images / 255.0
test_images = test_images / 255.0
plt.imshow(train_images[100], cmap="grey")
plt.colorbar()
```

```python
plt.figure(figsize=(10,10))
plt.subplots_adjust(hspace=0.5, wspace=0.5)
for i in range(25):
    plt.subplot(5,5,i+1)
    plt.imshow(train_images[i], cmap=plt.cm.binary)
    plt.xlabel(train_labels[i])
plt.show()
```

```python
model = tf.keras.models.Sequential([
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(128, activation=tf.nn.relu),
    tf.keras.layers.Dense(10, activation=tf.nn.softmax)
])
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.fit(train_images, train_labels, epochs=10)
```

```python
test_loss, test_acc = model.evaluate(test_images, test_labels, verbose=2)
```

```python
index = 100
plt.imshow(test_images[index], cmap=plt.cm.binary)
img_array = np.expand_dims(test_images[index], axis=0)
np.argmax(model.predict(img_array))
```
